```python
def oneDTravelWave(x,predict_interval,gamma, mass, ypsilon,tao=1, delta_x=1,delta_t_integration=0.1, delta_t_frame=5):
    """
    x........2D array
    """
    shape = x.shape
    result = np.zeros((shape[0],shape[1],predict_interval))
    second_order_deriv = np.array([1., -2., 1.]) / (delta_x*delta_x)
    x_up = x.copy()
    for i in range(predict_interval):
        num_frames = int(delta_t_frame/delta_t_integration)
        for j in range(num_frames):
            local_term = -gamma * x_up * (1.- x_up) *(0.5-x_up) + mass*x_up*(1-x_up) 
            tt = np.pad(x_up,((0,0),(1,1)),mode='edge')
            shape = tt.shape
            diffuse_term = ypsilon * (tt[:,0:(shape[1]-2)]-2*tt[:,1:(shape[1]-1)] +tt[:,2:(shape[1])] )/(delta_x*delta_x)
            x_up = x_up + (delta_t_integration / tao)*(local_term + diffuse_term)
        result[:,:,i] = x_up
    return result

def loss(paras, X, Y,tao,delta_x,delta_t_integration, delta_t_frame):
    """
    paras[0]: gamma
    paras[1]: mass
    paras[2]: ypsilon
    """
    gamma = paras[0]
    mass = paras[1]
    ypsilon = paras[2]
    shape = Y.shape
    predict_interval = shape[2]
    prediction = oneDTravelWave(X, predict_interval,gamma, mass, ypsilon,tao, delta_x,delta_t_integration,delta_t_frame)
    error = np.mean(np.square(Y-prediction))
    return error
    
def get_data(phase_fields, ids, predict_interval):
    X = []
    Y = []
    for i in ids:
        X.append(phase_fields[i])
        Y.append(phase_fields[(i+1):(i+1+predict_interval)])
    return np.array(X), np.swapaxes(np.array(Y),1,2)
```

```python
from keras import backend as K
from keras.layers import Layer

class OneDTravelWave_OneStep(Layer):

    def __init__(self, output_dim, delta_x,**kwargs):
        self.output_dim = output_dim
        #self.delta_x = delta_x
        var = np.array([1., -2., 1.]) / (delta_x*delta_x)
        var = np.expand_dims(var, axis=-1)  # 1 in channel
        var = np.expand_dims(var, axis=-1)  # 1 out channel 
        self.second_order_deriv = K.variable(var)
        super(OneDTravelWave_OneStep, self).__init__(**kwargs)

    def build(self, input_shape):
        # Create a trainable weight variable for this layer.
        self.tao = self.add_weight(name='tao', 
                                     shape=(1,),
                                     initializer='uniform',
                                     trainable=True)
        self.ypsilon  = self.add_weight(name='ypsilon', 
                                     shape=(1,),
                                     initializer='uniform',
                                     trainable=True)
        self.gamma = self.add_weight(name='gamma', 
                                     shape=(1,),
                                     initializer='uniform',
                                     trainable=True)
        self.mass = self.add_weight(name='mass', 
                                     shape=(1,),
                                     initializer='uniform',
                                     trainable=True)
        self.activation = keras.activations.linear
        super(OneDTravelWave_OneStep, self).build(input_shape)  # Be sure to call this at the end

    def call(self, x):
        local_term = -self.gamma * x * (1.- x) *(0.5-x) + self.mass*x*(1-x) 
        diffuse_term = self.ypsilon * K.conv1d(x, self.second_order_deriv,padding='same')
        return self.activation(x + self.tao*local_term + self.tao*diffuse_term)
    def compute_output_shape(self, input_shape):
        return (input_shape[0],) + self.output_dim
```

```python
def get_data(phase_fields, ids,predict_interval):
    X = []
    Y = []
    for i in ids:
        X.append(phase_fields[i])
        Y.append(phase_fields[(i+1):(i+1+predict_interval)])
    return np.expand_dims(np.array(X),axis=-1), np.array(Y)
```

```python
shape = phase_fields.shape
inputs = Input((shape[1],1))
one_step = OneDTravelWave_OneStep(output_dim=(shape[1],1), delta_x=distances[1]-distances[0])
```

```python
inputs.shape
```

```python
predict_all = []
for i in range(predict_interval):
    if i == 0:
        predict_all.append(one_step(inputs))
    else:
        predict_all.append(one_step(predict_all[-1]))
        #predict_all.append(one_step(Concatenate(axis=1)([Cropping3D(cropping=((0,i),(0,0),(0,0)))(inputs),]+predict_all)))
out0 = Concatenate(axis=-1)(predict_all)
out = Permute((2,1))(out0)
```

```python
from keras.optimizers import Adam
```

```python
model = Model(inputs=inputs,outputs=out)
#model.compile(loss='binary_crossentropy', optimizer='adam')
opt = Adam(learning_rate=0.0001)
model.compile(loss='mse', optimizer=opt)
```

```python
earlyStopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=500, verbose=0, mode='min')
mcp_save = keras.callbacks.ModelCheckpoint('simple_phase_field_model.hdf5', save_best_only=True, save_weights_only=True, monitor='val_loss', mode='min')
```
